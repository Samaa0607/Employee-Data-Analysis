```python
df.drop_duplicates(inplace=True)
```

```python
df = df.rename(columns={"Exit Date": "EXData" , "Annual Salary": "Annual_Salary", "Bonus %": "Bonus_%", "Business Unit": "Business_Unit" , "Job Title": "Job_Title" , "Full Name": "Full_Name", "Hire Date": "Hire_Date"})
df.head(5)
```

```python
df.drop(['EXData','EEID'] ,axis=1, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Bonus_%'] = df['Bonus_%'].str.replace('%', '').str.strip()
```

```python
df['Bonus_%'] = df['Bonus_%'].astype(int)
```

```python
df['Annual_Salary'] = df['Annual_Salary'].str.replace('$', '').str.strip()
```

```python
df['Annual_Salary'] = df['Annual_Salary'].str.replace(',', '').str.strip()
```

```python
df['Annual_Salary'] = df['Annual_Salary'].astype(int)
```

```python
df['Hire_Date'] = pd.to_datetime(df['Hire_Date'], errors='coerce')
```

```python
df['Hire_Year'] = df['Hire_Date'].dt.year
df['Hire_Month'] = df['Hire_Date'].dt.month
```
